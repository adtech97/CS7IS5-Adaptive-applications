```python
import numpy as np
from sklearn.metrics.pairwise import cosine_similarity




user_preferences = {
    'Type_Cardio': 0,  
    'Type_Strength': 1, 
    'Type_Stretching':0,
    'BodyPart_Abdominals': 0,  
    'BodyPart_Biceps': 1,  
    'BodyPart_Chest': 1,  
    'BodyPart_Forearms': 0,  
    'BodyPart_Neck': 0,  
    'BodyPart_Shoulders': 1,  
    'BodyPart_Triceps': 0,
    'Level_Beginner': 0,
    'Level_Expert': 0,
    'Level_Intermediate': 1,  
    'Equipment_Gym': 1,  
    'Equipment_Body_Only': 0,  
    'BodyPart_Legs': 0,  
    'BodyPart_Back': 0,  
    'BodyPart_FullBody': 0,  
}

cosine_similarities = []
user_vector = np.array(list(user_preferences.values())).reshape(1, -1)

for i in range(len(data_encoded)):
    exc_vector = data_encoded.iloc[i, 3:].values.reshape(1, -1)
    similarity = cosine_similarity(exc_vector, user_vector)
    cosine_similarities.append(similarity[0][0])
    
data_encoded['Cosine_Similarity'] = cosine_similarities

data_encoded_sorted = data_encoded.sort_values(by='Cosine_Similarity', ascending=False)

max_similarity = data_encoded_sorted['Cosine_Similarity'].max()

recommended_exercises_with_highest_similarity = data_encoded_sorted[
    data_encoded_sorted['Cosine_Similarity'] == max_similarity
]    
```

```python
recommended_exercises_with_highest_similarity
```

```python
shuffled_recommended_exercises = recommended_exercises_with_highest_similarity.sample(frac=1).reset_index(drop=True)

top_10_recommended_exercises = shuffled_recommended_exercises.head(10)
```

```python
top_10_recommended_exercises.drop(columns=['Cosine_Similarity'],inplace=True)
```

```python
user_selected_indices = [0, 3, 5]
user_selected_exercises = top_10_recommended_exercises.iloc[user_selected_indices]
```

```python
user_selected_exercises
#Must add to the database
unique_vectors_df = user_selected_exercises.iloc[:, 3:].drop_duplicates()
unique_vectors_df
```

```python

data_vectors = data_encoded.iloc[:, 3:].values
final_recommendations = []
for index, unique_vector_row in unique_vectors_df.iterrows():
    unique_vector = unique_vector_row.values.reshape(1, -1)
    similarities = cosine_similarity(unique_vector, data_vectors)[0]
    similarities_with_titles = pd.DataFrame({'Title': data_encoded['Title'], 'Similarity': similarities})
    sorted_similarities = similarities_with_titles.sort_values(by='Similarity', ascending=False)
    max_similarity_score = sorted_similarities['Similarity'].max()
    top_exercises = sorted_similarities[sorted_similarities['Similarity'] == max_similarity_score]
    random_selection = top_exercises.sample(n=min(3, len(top_exercises)), random_state=42)
    final_recommendations.append(random_selection['Title'].tolist())

```
